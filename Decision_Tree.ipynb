```python
#pandas library is used to read and analyze data
import pandas as pd #helps in Reading and writing data from various formats CSV, Excel, databases, etc.
```

```python
dataset=pd.read_csv("50_Startups.csv") #used to read data from CSV files into a Pandas,particularly for data preprocessing
#creating dataset as a variable
```

```python
dataset #execution result will be saved in this dataset
```

```python
#get_dummies helps convert categorical data into numerical using one-hot encoding
dataset=pd.get_dummies(dataset,drop_first=True)
#it helps in reducing extra column created during dummy variable creation
#drop_first,allows us to exclude the first column from the table
```

```python
dataset #execution result will be saved in this dataset
```

```python
#to get the names of the coloums we use this function
dataset.columns
```

```python
#splitting inputs and outputs, to find dependent variable
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
#dataset[[]] refers to selecting or extracting a column from a dataset to be used as the dependent variable in a model
```

```python
independent #input variable
```

```python
#splitting inputs and outputs, to find dependent variable
dependent=dataset[['Profit']]
#dataset[[]] refers to selecting or extracting a column from a dataset to be used as the dependent variable in a model
```

```python
#sklearn provides functions for splitting data into training and test sets, evaluating machine learning models, and performing cross-validation
from sklearn.model_selection import train_test_split #dot(.) denotes a file inside the sklearn library
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0) #train_test_split, Split Data Into Training and Testing Sets
#setting values for input(x) and output(y)
```

```python
#sklearn contain different functions for performing machine learning with linear models
from sklearn.tree import DecisionTreeRegressor #procedure
regressor=DecisionTreeRegressor(criterion='poisson', splitter='best') #DecisionTreeRegressor class is assigned to regressor variable,predicts the value of a dependent variable
regressor=regressor.fit(X_train, y_train) #fit function is substituting the values, it is used to train the model
#model,weight,bias are all saved under regressor only
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#Evaluation metrics
y_pred=regressor.predict(X_test) #contains all the predicted values for the input values in the X_test
#to cross check the values
```

```python
from sklearn.metrics import r2_score #provides functions for measuring model quality
r_score=r2_score(y_test,y_pred) #Measuring model fit using R2
```

```python
r_score #value is nearly to 1 means it is a good model
```
